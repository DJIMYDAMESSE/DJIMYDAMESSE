## Analyse exploratoire

## Pays avec le plus grand nombre de cas

```python
covid19.head()
```

```python
covid19_by_country =  covid19.groupby("countriesAndTerritories")["cases", "deaths"].sum().sort_values("cases", ascending=False)
covid19_by_country 
```

**Remarque:** *on conclut que les états unis sont le pays avec le plus grand nombre de cas en 2020 qui est de* **16256754** *pour un nombre de décès de* **299177**

## pays avec le taux de mortalité le plus élévé

```python
covid19_by_country["taux_de_mortalité"] = covid19_by_country["deaths"] / covid19_by_country["cases"]
covid19_by_country 
```

```python
plt.figure(figsize=(14, 12))
ax = covid19_by_country["taux_de_mortalité"].sort_values(ascending=False).head(20).plot(kind="bar")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="left")
ax.set_xlabel("Country")
ax.set_ylabel("taux_de_mortalité")
ax.set_title("pays avec le taux de mortalité élévé")
plt.savefig("pays_avec_le_taux_de_mortalité_élévé.pdf")
plt.show()
```

**Remarque:** *on peut ainsi dire que le pays avec le plus grand taux de mortalité est le yémem qui est de plus de* **0.27%**

## Pays avec le plus de morts

```python
plt.figure(figsize=(10, 6))
ax = covid19_by_country["deaths"].sort_values(ascending=False).head(20).plot(kind="barh")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="left")
ax.set_xlabel("Country")
ax.set_ylabel("Nombre de mort")
ax.set_title("Pays avec le plus de morts de la covid19")
plt.savefig("Pays_avec_le_plus_de_morts_de_la_covid19.pdf")
plt.show()
```

## Effet du confinemenz sur le nimbre de cas 

```python
covid19_by_month = covid19.groupby("month")["cases", "deaths"].sum()
```

```python
covid19_by_month.head()
```

```python
fig=plt.figure(figsize=(15, 10))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

covid19_by_month["cases"].plot(kind="line", ax=ax1)
ax1.set_title("Nombre total de cas de covid par mois")
ax1.set_label("mois")
ax1.set_ylabel("nombre de cas")
covid19_by_month["deaths"].plot(kind="line", ax=ax2)
ax2.set_title("Nombre total de morts de covid par mois")
ax2.set_label("mois")
ax2.set_ylabel("nombre de morts")
```

**Remarque:** *de facon globale , on constate une diminution une décroissance du nombre de morts et de cas de covid. essayons de voir si cela est egalement effectif dans les pays ou le confinement a été tres stict*

```python
covid19_germany = covid19[covid19.countriesAndTerritories == "Germany"]
covid19_germany_monthwise = covid19_germany.groupby("month")["cases", "deaths"].sum()
covid19_germany_grouped = covid19_germany_monthwise.reset_index()

covid19_uk = covid19[covid19.countriesAndTerritories == "United_Kingdom"]
covid19_uk_monthwise = covid19_uk.groupby("month")["cases", "deaths"].sum()
covid19_uk_grouped = covid19_uk_monthwise.reset_index()

covid19_france = covid19[covid19.countriesAndTerritories == "France"]
covid19_france_monthwise = covid19_france.groupby("month")["cases", "deaths"].sum()
covid19_france_grouped = covid19_france_monthwise.reset_index()

covid19_italy = covid19[covid19.countriesAndTerritories == "Italy"]
covid19_italy_monthwise = covid19_italy.groupby("month")["cases", "deaths"].sum()
covid19_italy_grouped = covid19_italy_monthwise.reset_index()
```

```python
fig=plt.figure(figsize=(20, 15))
            
ax1=fig.add_subplot(2,2,1)
covid19_uk_grouped.plot(kind="line",x="month", y="cases", ax=ax1)
ax1.set_title("Evolution du covid en UK")
               
ax2=fig.add_subplot(2,2,2)
covid19_germany_grouped.plot(kind="line",x="month", y="cases", ax=ax2)
ax2.set_title("Evolution du covid en Germany")
            
ax3=fig.add_subplot(2,2,3)
covid19_france_grouped.plot(kind="line",x="month", y="cases", ax=ax3)
ax3.set_title("Evolution du covid en France")

ax4=fig.add_subplot(2,2,4)
covid19_italy_grouped.plot(kind="line",x="month", y="cases", ax=ax4)
ax4.set_title("Evolution du covid en Italy")

```
